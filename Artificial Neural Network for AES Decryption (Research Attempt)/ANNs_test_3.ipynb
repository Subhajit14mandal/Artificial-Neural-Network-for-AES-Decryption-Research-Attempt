```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv('ANNs_file_Encrypt_by_One_Key.csv')
data.head()
```

```python
def text_to_vector(text):
  return np.array([ord(char)/128 for char in text])
```

```python
# Calculate max_length considering both Plain_text and Cipher_text columns
max_length=max(data['Plain_text'].apply(len).max(),data['Cipher_text'].apply(len).max())
max_length
```

```python
# Apply padding to both columns with the adjusted max_length
data['Plain_text']=data['Plain_text'].apply(lambda x: np.pad(text_to_vector(x),(0,max_length-len(x)),'constant'))
data['Cipher_text']=data['Cipher_text'].apply(lambda x: np.pad(text_to_vector(x),(0,max_length-len(x)),'constant'))
data.head()
```

```python
y=np.stack(data['Plain_text'].values)
x=np.stack(data['Cipher_text'].values)
```

```python
ss=StandardScaler()
x=ss.fit_transform(x)
y=ss.fit_transform(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=Sequential()
model.add(Dense(256,activation='relu',input_dim=x_train.shape[1]))
# model.add(Dense(192,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(max_length,activation='linear'))
```

```python
model.compile(optimizer='adam',loss='mean_squared_error',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=100,batch_size=10,verbose=1)
loss,accuracy=model.evaluate(x_test,y_test)
accuracy
```
